```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
yoe=pd.DataFrame(faamgeduonly['yearsofexperience'])
yoc=pd.DataFrame(faamgeduonly['totalyearlycompensation'])

yoe_train, yoe_test, yoc_train, yoc_test = train_test_split(yoe, yoc, test_size = 0.20)
linregYOE=LinearRegression()
linregYOE.fit(yoe_train,yoc_train)

regLineX=yoe_train
regLineY=linregYOE.intercept_+linregYOE.coef_*yoe_train

f = plt.figure(figsize=(16, 8))
plt.scatter(yoe_train, yoc_train)
plt.plot(regLineX, regLineY, 'r-', linewidth = 3)
plt.show()
```

```python
yoc_train_pred = linregYOE.predict(yoe_train)

f = plt.figure(figsize=(16, 8))
plt.scatter(yoe_train, yoc_train, color = "g")
plt.scatter(yoe_train, yoc_train_pred, color = "r")
plt.show()
```

```python
# Explained Variance (R^2)
print("Explained Variance (R^2) \t:", linregYOE.score(yoe_train, yoc_train))

# Mean Squared Error (MSE)
def mean_sq_err(actual, predicted):
    '''Returns the Mean Squared Error of actual and predicted values'''
    return np.mean(np.square(np.array(actual) - np.array(predicted)))


mse = mean_sq_err(yoe_train, yoc_train_pred)
print("Mean Squared Error (MSE) \t:", mse)
print("Root Mean Squared Error (RMSE) \t:", np.sqrt(mse))
```

```python
yoc_test_pred = linregYOE.predict(yoe_test)

f = plt.figure(figsize=(16, 8))
plt.scatter(yoe_test, yoc_test, color ="g")
plt.scatter(yoe_test, yoc_test_pred, color = "r")
plt.show()
```

```python
# Explained Variance (R^2)
print("Explained Variance (R^2) \t:", linregYOE.score(yoe_test, yoc_test))

# Mean Squared Error (MSE)
def mean_sq_err(actual, predicted):
    '''Returns the Mean Squared Error of actual and predicted values'''
    return np.mean(np.square(np.array(actual) - np.array(predicted)))


mse = mean_sq_err(yoe_test, yoe_test_pred)
print("Mean Squared Error (MSE) \t:", mse)
print("Root Mean Squared Error (RMSE) \t:", np.sqrt(mse))
```

```python
# Predict Response corresponding to Predictors
yoc_train_pred = linregYOE.predict(yoe_train)
yoc_test_pred = linregYOE.predict(yoe_test)
    
# Plot the Predictions vs the True values
f, axes = plt.subplots(1, 2, figsize=(24, 12))
axes[0].scatter(yoe_train, yoc_train, color = "blue")
axes[0].plot(regLineX, regLineY, 'r-', linewidth = 3)
axes[0].set_title("Regression Line on Train Set")
axes[1].scatter(yoe_test, yoc_test, color = "green")
axes[1].scatter(yoe_test, yoc_test_pred, color = "red")
axes[1].set_title("Model Prediction on Test Set")
plt.show()

# Check the Goodness of Fit (on Train Data)
print("Goodness of Fit of Model \tTrain Dataset")
print("Explained Variance (R^2) \t:", linregYOE.score(yoe_train, yoc_train))
print("Mean Squared Error (MSE) \t:", mean_squared_error(yoc_train, yoc_train_pred))
print()

# Check the Goodness of Fit (on Test Data)
print("Goodness of Fit of Model \tTest Dataset")
print("Explained Variance (R^2) \t:", linregYOE.score(yoe_test, yoc_test))
print("Prediction Accuracy of Model \tTest Dataset")
print("Mean Squared Error (MSE) \t:", mean_squared_error(yoc_test, yoc_test_pred))
print()
```
